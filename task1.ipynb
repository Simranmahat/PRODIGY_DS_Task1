```python
#Data Visualization
#PLOTTING HISPLOT FROM 2001 TO 2022


cols={'2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2012','2013','2014','2015','2016','2017','2018'
      ,'2019','2020','2021','2022'}
```

```python
import matplotlib.pyplot as plt
for i in cols:
    fig=plt.figure(figsize=(5,5))
    plt.hist(df[i],color=	'#355E3B',bins=10)
    plt.xlabel(i)
    plt.show()
```

```python
#Extract the years (columns)
years=df.columns[1:]

#Sum the values for each year across all the countries
total_values=df[years].sum()

#Plotting the bar chart
plt.figure(figsize=(30,30))
plt.barh(years,total_values,color='#355E3B')
plt.xlabel('Total Values')
plt.ylabel('Year',size=20)
plt.title('Total Values per Year',size=15)
plt.show()
```
